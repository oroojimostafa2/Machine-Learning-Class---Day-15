```python
def dataformat(df, wellbore_name):
    """
    Funtion for resetting the index for each well bore and converting
    the date column to datetime format
    
    inputs ->
    
    df = Dataframe of field
    wellbore_name = wellbore name whose data we want
    
    return 
    
    df_well = dataframe of the well that we want, with resetted index
    and datetime format dateprd
    """
    
    df_well = df[df["NPD_WELL_BORE_NAME"] == wellbore_name]
    df_well = df_well.reset_index()
    df_well = df_well.drop("index", axis=1)
    df_well["DATEPRD"] = pd.to_datetime(df_well["DATEPRD"])
    
    return df_well
```

```python
df_f12_function = dataformat(df, "15/9-F-12")
df_f12_function.info()
```
